```python

file_name =  "catCFDI_V_4_16112023.xls"



from xlrd import open_workbook, Book as bookType
from typing import Union
import json

# book = open_workbook(file_name,on_demand=True)

file_name = "catCFDI_V_4_16112023.xls"


# We should set the encoding to utf-8
book: bookType = open_workbook(file_name, on_demand=True, encoding_override="utf-8")
sheets_ignore = ["c_TasaOCuota", "c_NumPedimentoAduana", "c_Aduana", "c_PatenteAduanal"]

for name in book.sheet_names():
    if name in sheets_ignore:
        continue
    print(name)
    sheet = book.sheet_by_name(name)

    startsaving = False
    headers = []
    data = []
    # We would find the row where the first cell value is equal to name.
    # We would then iterate and save the row values until we get to the end of the sheet or we find a row with an empty first cell.
    for row in sheet.get_rows():
        # If the sheet name starts with the row value
        if len(str(row[0].value)) > 0 and (name.lower()).startswith(str(row[0].value).lower()):
            print("Found the start of the sheet", name, row[0].value)
            headers = [cell.value for cell in row]
            headers[0] = "id"
            data.append(headers)
            startsaving = True
            # How I can know if the row is a merge cell?
            print(row[0].ctype)
            continue
        if startsaving:
            rowData = [cell.value for cell in row]
            # rowData = {}
            # for index, cell in enumerate(row):
            #     rowData[headers[index]] = cell.value
            data.append(rowData)
    


    # We should save the data content to a file called name.json
    filename_json = name + ".json"
    content_json = json.dumps(data, indent=4)
    with open(filename_json, "w") as f:
        f.write(content_json)

    # We should unload the sheet from memory
    book.unload_sheet(name)
```
