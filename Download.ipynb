```python
import requests
from bs4 import BeautifulSoup

# Make a GET request to the page
url = "http://omawww.sat.gob.mx/tramitesyservicios/Paginas/anexo_20.htm"  # Replace with the actual URL of the page
response = requests.get(url)

# Check if the request was successful
if response.status_code != 200:
    print("Request to {} failed with status code {}".format(url, response.status_code))
    exit()

# Debug the response
# print("Response content:")
# print(response.content)

# Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Debug the parsed content with pretty print
# print("Parsed content:")
# print(soup.prettify())

# Find the link with text "Catalogos"
link = soup.find("a", text="Catálogos")

# for i in link:
#     print(i.text, i["href"])

# Get the URL of the link
if link:
    catalogos_url = link["href"]
    print("URL for 'Catalogos' link:", catalogos_url)

    # Lets resolve the relative catalogos_url to absolute URL based on the original URL
    from urllib.parse import urljoin
    catalogos_url = urljoin(url, catalogos_url)

    # Make a GET request to the link
    response = requests.get(catalogos_url)

    # Check if the request was successful
    if response.status_code != 200:
        print("Request to {} failed with status code {}".format(catalogos_url, response.status_code))
        exit()

    # Get the filename from the catalogos_url
    import os
    filename = os.path.basename(catalogos_url)

    # Save the file to disk
    with open(filename, "wb") as f:
        f.write(response.content)
        print("File saved to disk as 'catalogos.zip'")

else:
    print("Link 'Catalogos' not found on the page")
```
